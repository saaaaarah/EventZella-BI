```python
if 'Date' not in df_res.columns:
    print("'Date' is already the index or not present in the DataFrame.")
else:
    df_res.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))
plot_acf(df_res['final_price'], lags=40)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF) Plot')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_res['final_price'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(df_res['final_price'], model='additive')
decomp.plot()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model_arima = ARIMA(df_res['final_price'], order=(1,1,1))
model_arima_fit = model_arima.fit()

forecast_arima = model_arima_fit.forecast(steps=6)
print(forecast_arima)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model_sarima = SARIMAX(
    df_res['final_price'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

model_sarima_fit = model_sarima.fit()

forecast_sarima = model_sarima_fit.forecast(steps=6)
print(forecast_sarima)
```

```python
train = df_res.iloc[:-6]
test = df_res.iloc[-6:]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# ARIMA predictions
pred_arima = model_arima_fit.forecast(steps=6)

mae_arima = mean_absolute_error(test['final_price'], pred_arima)
rmse_arima = np.sqrt(mean_squared_error(test['final_price'], pred_arima))
mape_arima = np.mean(np.abs((test['final_price'] - pred_arima) / test['final_price'])) * 100

print(mae_arima, rmse_arima, mape_arima)
```

```python

pred_sarima = model_sarima_fit.forecast(steps=6)

mae_sarima = mean_absolute_error(test['final_price'], pred_sarima)
rmse_sarima = np.sqrt(mean_squared_error(test['final_price'], pred_sarima))
mape_sarima = np.mean(np.abs((test['final_price'] - pred_sarima) / test['final_price'])) * 100

print(mae_sarima, rmse_sarima, mape_sarima)
```

```python
comparison = pd.DataFrame({
    'Model': ['ARIMA', 'SARIMA'],
    'MAE': [mae_arima, mae_sarima],
    'RMSE': [rmse_arima, rmse_sarima],
    'MAPE': [mape_arima, mape_sarima]
})

print(comparison)
```
